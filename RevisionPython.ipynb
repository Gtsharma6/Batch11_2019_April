```python
def prime(num,check=2) :
    if check == num // 2 + 1 : 
        return True 
    elif num % check == 0 : 
        return False
    return prime(num,check+1)
    
```

```python
prime1 = lambda num,check=2 : True if check == num // 2 + 1 else False if num % check == 0 else prime(num,check+1) 
print(prime1(121))
print(prime1(127))
```
